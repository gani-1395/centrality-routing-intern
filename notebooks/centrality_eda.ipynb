```python
import matplotlib.pyplot as plt
import networkx as nx

# 1. Regenerate your 20-node random graph
G = nx.erdos_renyi_graph(20, 0.3, seed=42)

# 2. Calculate the 4 primary network centrality metrics
degree_cent = nx.degree_centrality(G)
between_cent = nx.betweenness_centrality(G)
closeness_cent = nx.closeness_centrality(G)
eigen_cent = nx.eigenvector_centrality(G)

print("Node 0 Metrics Snapshot:")
print(f"- Degree Centrality: {degree_cent[0]:.3f}")
print(f"- Betweenness Centrality: {between_cent[0]:.3f}")
print(f"- Closeness Centrality: {closeness_cent[0]:.3f}")
print(f"- Eigenvector Centrality: {eigen_cent[0]:.3f}\n")

# 3. Visualize: Color code the nodes based on their Betweenness Score
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)

nodes = nx.draw_networkx_nodes(
    G,
    pos,
    node_color=list(between_cent.values()),
    cmap=plt.cm.plasma,
    node_size=600,
)
nx.draw_networkx_edges(G, pos, edge_color="gainsboro")
nx.draw_networkx_labels(G, pos, font_color="white", font_weight="bold")

plt.colorbar(nodes, label="Betweenness Centrality Score")
plt.title("Network Node Importance via Betweenness Centrality")
plt.axis("off")

os.makedirs("results", exist_ok=True)
plt.savefig("results/centrality_analysis.png", bbox_inches="tight")
plt.show()
```
